# 쿼리에 따른 데이터 검색

```python
import os, requests
from dotenv import load_dotenv
import pandas as pd

load_dotenv()

queries = ['Architectural Urbanism', 'Melbourne Seoul', 'KTA projects', 'public architecture', 'civic architecture']

API_KEY = os.getenv("DATAON_SEARCH_API_KEY")
assert API_KEY and API_KEY.strip(), "환경변수(DATAON_API_KEY)가 비어있어요!"

url = "https://dataon.kisti.re.kr/rest/api/search/dataset/"

df = pd.DataFrame()
for query in queries:
    params = {"key": API_KEY, "query": query, "from": 0, "size": 5}
    # key / CHAR / 필수 / API_KEY
    # query / CHAR / 필수 / 검색키워드
    # from / CHAR / 옵션 / 페이지시작위치
    # size / CHAR / 옵션 / 페이지사이즈

    res = requests.get(url, params=params, timeout=20)
    data = res.json()
    print(data['response'])
    if "records" in data:
        tmp = pd.DataFrame(data["records"])
        tmp["query"] = query
        df = pd.concat([df, tmp], ignore_index=True)

display(df)
df.to_csv('search_results.csv', index=False, encoding='utf-8')
```
